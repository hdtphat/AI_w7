```python
# Thêm datasheet
import pandas as pd
url = "https://raw.githubusercontent.com/hdtphat/Linear-Regression/main/Datasheets/50_Startups.csv"
data = pd.read_csv(url)
data.head

###
import seaborn as sns
import matplotlib.pyplot as plt
data.hist()
plt.show()

###
correlation = data.corr()
print(correlation)

###
sns.pairplot(data)
```

```python
# Bản đồ nhiệt
sns.heatmap(correlation, annot=True)
```

```python
from sklearn import linear_model
model = linear_model.LinearRegression()
model.fit(X=pd.DataFrame(data['R&D Spend']), y=data['Profit'])
print('Interscept : ', model.intercept_)
print('Coefficent : ', model.coef_)
model.score(X=pd.DataFrame(data['R&D Spend']), y=data['Profit'])
```

```python
# vẽ đồ thị hồi quy bậc 1
data.plot(kind='scatter', x='R&D Spend', y='Profit', figsize=(9,9), color='black')
y_predict = model.predict(X=pd.DataFrame(data['R&D Spend']))
plt.plot(data['R&D Spend'], y_predict, color='blue')
```

```python
# vẽ đồ thị hồi quy bậc 2
model = linear_model.LinearRegression()
X = pd.DataFrame([data['R&D Spend'], data['R&D Spend']**2]).T
y = data['Profit'] #lúc đc lúc k
model.fit(X, y=data['Profit'])
print('Intercept:', model.intercept_)
print('Coeffient:', model.coef_)
print('độ chính xác:', model.score(X, y))

import numpy as np
range = np.arange(5000, 200000, 1000) #đường màu đỏ chạy từ x=5000 đến x=200000, bước chạy 1000
X_range = pd.DataFrame([range, range**2]).T
Y_predict = model.predict(X_range)
data.plot(kind='scatter', x='R&D Spend', y='Profit', figsize=(9,9), color='black', xlim=(0.7))
plt.plot(range, Y_predict, color='red')
```

```python
# vẽ đồ thị hồi quy bậc 10
model = linear_model.LinearRegression()
X = pd.DataFrame([data['R&D Spend'], data['R&D Spend']**2, data['R&D Spend']**3, data['R&D Spend']**4, data['R&D Spend']**5, data['R&D Spend']**6, data['R&D Spend']**7, data['R&D Spend']**8, data['R&D Spend']**9, data['R&D Spend']**10]).T
y = data['Profit']
model.fit(X, y=data['Profit'])
print('Intercept:', model.intercept_)
print('Coeffient:', model.coef_)
print('độ chính xác:', model.score(X, y))

import numpy as np
range = np.arange(5000, 200000, 1000)
X_range = pd.DataFrame([range, range**2, range**3, range**4, range**5, range**6, range**7, range**8, range**9, range**10]).T
Y_predict = model.predict(X_range)
data.plot(kind='scatter', x='R&D Spend', y='Profit', figsize=(9,9), color='black', xlim=(0.7))
plt.plot(range, Y_predict, color='red')
```
