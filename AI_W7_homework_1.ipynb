```python

import pandas as pd
url='https://raw.githubusercontent.com/hdtphat/AI_w7/main/Datasheets/classification.csv'
data=pd.read_csv(url)
data.head()

import matplotlib.pyplot as plt
data.hist()
plt.show()

correlation=data.corr()
print(correlation)
```

```python
import seaborn as sns
sns.heatmap(correlation,annot=True)
sns.pairplot(data)
```

```python
from sklearn import linear_model
model=linear_model.LinearRegression()
#learning data
model.fit(X=data.loc[:,['age','interest']],y=data['success'])
print('intercept: ',model.intercept_)
print('coeffient: ',model.coef_)
print('Accuracy: ',model.score(X=data.loc[:,['age','interest']],y=data['success']))

model=linear_model.LinearRegression()
X=pd.DataFrame([data['age'],data['age']**2,data['age']**3]).T 
model.fit(X,y=data['success'])
print("Intercept 3rd order:", model.intercept_)
print("Coeffient 3rd order:", model.coef_)
print('Accuracy 3rd order:',model.score(X,y=data['success']))
```

```python
import numpy as np
range= np.arange(0,50,1)
X_range=pd.DataFrame([range,range**2,range**3]).T 
y_pred=model.predict(X=X_range)
data.plot(kind='scatter',x='age',
          y='success',figsize=(9,9),color='black',
          xlim=(0,50))
plt.plot(range,y_pred,color='black')
```
